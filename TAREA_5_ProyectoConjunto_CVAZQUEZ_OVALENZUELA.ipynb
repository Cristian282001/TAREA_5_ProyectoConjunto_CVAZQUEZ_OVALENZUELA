## Ejercicio 1:

> 1. Usar como función que mayora a $f(x)$ a $t(x)=a \sin(\pi x)$ donde a es el máximo de la función $f(x)$ y graficarlas en una misma gráfica, para validar que en realidad si cumple la condición $t(x)\geq f(x)$.

### Código de solución estudiante 1

```python
# Libreria

import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import pandas as pd

#  Función lambda "fx" que calcula el valor de la función 
f=lambda x: 30*(x**2-2*x**3+x**4)

# Calcula el valor máximo de la función "f(x)" 
# usando el método de optimización
max_f=f(optimize.fmin(lambda x:-f(x), 0, disp=False))
print("El máximo de f(x) es ", max_f[0])

# Función lambda "tx" que calcula el valor de la función matemática "t(x)"
t=lambda x: max_f*np.sin(np.pi*x)

# Crea un arreglo de valores de "x" desde 0 hasta 1
x = np.arange(0,1,0.01)

# Grafica 
plt.title('Funciones')
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x, f(x), color = "green", label = "f(x)")
plt.plot(x, t(x), color = "orange", label = "t(x)")
plt.legend(loc = "best")
plt.grid()
```

### Interpretacion

-  Este código tiene como objetivo trazar dos funciones matemáticas `"f(x)"` y `"t(x)"` utilizando la biblioteca Matplotlib. Primero, se define la función `"f(x)"` utilizando una expresión lambda y se utiliza el método de optimización scipy.optimize.fmin para encontrar su valor máximo. A continuación, se define la función `"t(x)"` utilizando el valor máximo de `"f(x)"` calculado previamente y se crea un arreglo de valores de `"x"` utilizando la biblioteca NumPy. Luego, se utiliza la función `plt.plot` de Matplotlib para trazar ambas funciones en una figura, con sus correspondientes etiquetas y colores. Finalmente, se agrega una leyenda y una cuadrícula en la figura para mejorar su legibilidad. En resumen, este código ilustra cómo trazar funciones matemáticas utilizando Python y la biblioteca Matplotlib.

- 4. Según el punto 3, generar 10000 puntos aleatorios que sigan la distribución $g(x)$ y comparar con su histograma para validar que los puntos generados siguen la distribución deseada. El resultado debe ser como sigue:


### Código de solución estudiante 1

```python
# Libreria

import numpy as np
import matplotlib.pyplot as plt
from typing import Optional
```

```python
# Esta es una función sinusoidal escalada, se utiliza para generar números aleatorios
def g(x):
    y = (np.pi*np.sin(np.pi*x))/2  #sin
    return y
```

```python
# Esta función se utiliza para transformar los números aleatorios generados
def g_inv(u):
    x = (np.arccos(-2*u+1))/(np.pi)
    return x
```

```python
# Devuelve una lista de los números aleatorios transformados
def m_aleatorias(n):
    u = np.random.uniform(0, 1, n)
    r = g_inv(u)
    return r
```

```python
# Almacena la lista 
r1 = m_aleatorias(10000)

#Grafica
plt.figure(figsize=(10, 7))
plt.hist(r1, bins=50, density=True, range=(0, 1), color='yellow')
plt.plot(np.arange(0, 1, 0.01), g(np.arange(0, 1, 0.01)), 'blue')
plt.grid();
```

## Ejercicio 5:

- 5. Genere 500 puntos aleatorios usando el método de aceptación y rechazo y las funciones $f(x)$ y $t(x)$ para validar que todos los cálculos anteriores están correctamente realizados. El resultado debe de ser como sigue:


### Código de solución estudiante 1

```python
# Define la función y toma dos arreglos de números aleatorios y dos funciones 

def Acep_rechazo(R2,R1,f,t):
    f_x = f(R1)
    t_x = t(R1)
    
    # Contador 
    q = 0
    
    # Condición que compara
    condition = R2*t_x <=f_x
    
    #  Evalúa la condición definida anteriormente
    for i in range(len(R1)):
        if condition[i]:
            plt.plot(R1[i],R2[i]*t_x[i],'ob')
            q += 1
        else:
            plt.plot(R1[i],R2[i]*t_x[i],'o')
    plt.show()
    return q
```

```python
# Define EL arreglo
x = np.arange(0,1,0.01)
```

```python
#Grafica
plt.plot(x,f(x),label='f(x)')
plt.plot(x,t(x),label='t(x)')
plt.legend(loc = "best")
plt.grid();
```

```python
# Número de muestras aleatorias 
N = 500 
```

```python
# Se guarda la cantidad de muestras
R2 = np.random.rand(N)
R1 = m_aleatorias(N)
quantity_sin = Acep_rechazo(R2,R1,f,t)
```

```python
# Grafica completa

plt.plot(x,f(x),label='f(x)')
plt.plot(x,t(x),label='t(x)')
plt.legend(loc = "best")
plt.grid();

# Número de muestras aleatorias 
N = 500 

# Se guarda la cantidad de muestras
R2 = np.random.rand(N)
R1 = m_aleatorias(N)
quantity_sin = Acep_rechazo(R2,R1,f,t)
```
